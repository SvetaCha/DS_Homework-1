### Завдання 6 

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random
import math
import statistics

#генеруємо тисячу чисел в діапазоні від 0 до 100
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]
def minimum(x):
    minimum = x[0]
    for i in x:
        if i < minimum:
            minimum = i
    return minimum
def maximum(x):
    maximum = x[0]
    for i in x:
        if i > maximum:
            maximum = i
    return maximum
print(f'мінімальне: {minimum(list_with_random_numbers)}, максимальне: {maximum(list_with_random_numbers)}')
print('Проверка: ', min(list_with_random_numbers), max(list_with_random_numbers))
def avg(x):
    s = 0
    for i in x:
        s = s + i
    return s / len(x)
print('середнє значення послідовності:', avg(list_with_random_numbers))
print('Проверка:', statistics.mean(list_with_random_numbers))
def mediana(x):
    x.sort()
    if len(x) % 2 == 0:
        mediana = (x[len(x)//2 - 1] + x[len(x)//2]) / 2
    else:
        mediana = x[len(x)//2]
    return mediana
print('медіана послідовності:', mediana(list_with_random_numbers))
print('Проверка:', statistics.median(list_with_random_numbers))
def add_dict(my_list):
    my_dict = {}
    for i in my_list:
        if i in my_dict:
            my_dict[i] = my_dict[i] + 1
        else:
            my_dict[i] = 1
    return my_dict
freq_dict = add_dict(list_with_random_numbers)
print('частотa кожної цифри в послідовності', freq_dict)
def stdev(l):
    av_num = sum(l) / len(l)
    s = 0
    for i in l:
        s = s + (i - av_num) ** 2
    return (s / (len(l) - 1)) ** (1/2)
print('середнє квадратичне відхилення:', stdev(list_with_random_numbers))
print('Проверка:', statistics.stdev(list_with_random_numbers))
```
